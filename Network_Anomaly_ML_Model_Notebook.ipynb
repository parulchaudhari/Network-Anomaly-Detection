## UnderSampling

in this section, we have undersampled our data for majority of classes(class18, class10 and normal), keeping it to 50000 rows. 

Though you might think, we lost the huge data, but keeping whole data would lead to overfitting as it will majorly classify these classes over minority classes. 

```python
#creating duplcates for class 19 and 12 as we have very few samples like 2 or 3, else it will not work properly while over sampling
train_ = pd.DataFrame(np.repeat(train[train[42].isin(['class19','class12'])].values,2,axis=0))
train_.columns = train.columns
train_.shape
```

```python
train = pd.concat([train,train_],axis=0) #Merging duplicates class rows in train data
```

```python
#creating sub dataframes of majority classes and minority classes for undersampling process
train_class10 = train[train[42].isin(['class10'])]  
train_normal  =train[train[42].isin(['normal'])]
train_class18 = train[train[42].isin(['class18'])]
 
train_rem = train[~train[42].isin(['class10','normal','class18'])] #minority classes sub set
train_rem.shape, train_class10.shape, train_normal.shape, train_class18.shape, train.shape
```

```python
train = pd.concat([train_class10[:50000],train_normal[:50000],train_class18[:50000]], axis=0)  #keeping only 50000,50000 from the majority classes 
train = pd.concat([train, train_rem], axis=0)
train.shape
```


# Feature Selection

 Will use a combination of Chi, Anova and Mutual Information Statistics on numerical data, and then combine with our catgorical data while training the models for predictions


```python
# create x and y train
X = train.drop(columns=[0,1,2,42], axis=1) #dropping catg variable
y = train[[42]]

from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2
from sklearn.feature_selection import mutual_info_classif 
from sklearn.feature_selection import f_classif

def FeatureSelection(featureTest):
    bestfeatures = SelectKBest(score_func=featureTest, k='all')
    fit = bestfeatures.fit(X,y)
    dfscores = pd.DataFrame(fit.scores_)
    dfcolumns = pd.DataFrame(X.columns)
    featureScores = pd.concat([dfcolumns,dfscores],axis=1)
    featureScores.columns = ['Specs','Score']
    return featureScores
```

```python
X.head(1)
```

```python
%%time
#running the tests at once and storing them into different variables
featureScores = FeatureSelection(chi2)
MutualFeatures = FeatureSelection(mutual_info_classif)
AnovaFeatures = FeatureSelection(f_classif)
```

```python
#keeping only 5 top features from each test
num_F = 5
A = set(featureScores.nlargest(num_F,'Score')['Specs'])
B = set(MutualFeatures.nlargest(num_F,'Score')['Specs'])
C = set(AnovaFeatures.nlargest(num_F,'Score')['Specs'])
```

```python
#union of all test results
FinalFeatures = A | B | C 
pd.DataFrame(list(FinalFeatures),columns=['Most Important Features'])
```

```python
N_features = list(FinalFeatures) #converting set to list
```

# SMOTE - Oversample data

balancing our dataset as it's highly imbalance. 

```python
%%time
from imblearn.over_sampling import SMOTE
from collections import Counter
from matplotlib import pyplot

n_class_sample=45000
n_class_sample_major=45000

strategy = {0:n_class_sample, 1:n_class_sample, 2:n_class_sample, 3:n_class_sample, 4:n_class_sample, 5:n_class_sample,
            6:n_class_sample, 7:n_class_sample, 8:n_class_sample, 9:n_class_sample_major, 10:n_class_sample, 11:n_class_sample,
            12:n_class_sample, 13:n_class_sample, 14:n_class_sample, 15:n_class_sample, 16:n_class_sample, 17:n_class_sample_major,
            18:n_class_sample, 19:n_class_sample, 20:n_class_sample, 21:n_class_sample, 22:n_class_sample_major,
           }
           
oversample = SMOTE(sampling_strategy=strategy,k_neighbors=1)
X_train, y_train = oversample.fit_resample(X_train, y_train)

# summarize distribution
counter = Counter(y_train)
for k,v in counter.items():
	per = v / len(y_train) * 100
	print('Class=%d, n=%d (%.3f%%)' % (k, v, per))
# plot the distribution
pyplot.bar(counter.keys(), counter.values())
pyplot.show()
```

PCA also helped me in decreasing the time duration of training our model otherwise knn is very slow.


## Feature selection using PCA

```python
#PCA for faster execution and less number of features to decrease complexity of knn
from sklearn.decomposition import PCA
pca = PCA(random_state=42)
pca.fit(X_train)
plt.bar(range(1,len(pca.explained_variance_ratio_)+1), pca.explained_variance_ratio_)
plt.show()

var_cumu = np.cumsum(pca.explained_variance_ratio_)
fig = plt.figure(figsize=[12,8])
plt.vlines(x=15, ymax=1, ymin=0, colors="r", linestyles="--")
plt.hlines(y=0.95, xmax=30, xmin=0, colors="g", linestyles="--")
plt.plot(var_cumu)
plt.ylabel("Cumulative variance explained")
plt.show()

from sklearn.decomposition import IncrementalPCA
pca_final = IncrementalPCA(n_components=30) # after observing variance ratio, keeping 30 components for training the model 
```

keping 30 features as they provide us the 100% variance for our dataset

## K value Tuning
